```python
# load image from url
import requests
def load_image(url):
    return Image.open(requests.get(url, stream=True).raw)
```

```python
im2.rotate(90, fillcolor='white' ,expand=True)
```

```python
w,h = im2.width, im2.height
img = im2.resize((w*2, h*2)) #reducing size by 50%
img
```

```python
# enhancing color
en = ImageEnhance.Color(img)
```

```python
for i in range(-5, 6):
    en.enhance(i).save(f'img/img_enhance_{i}.jpg')
```

```python
# image drawing
dr = ImageDraw.Draw(img)
dr.circle((143, 219), 40, fill = 'red')
dr.circle((143, 219), radius = 80, outline = 'yellow', width = 5)
img
```

```python
from PIL import ImageFont
font = ImageFont.truetype(r'c:\Windows\Fonts\LATINWD.TTF', 25)
dr.text((143,219), ' target found', fill = 'black', font=font, stroke_fill= 'white', stroke_width= 1, anchor='ms')
start_point = (0, img.height//2)
end_point = (img.width, img.height//2)
dr.line(xy = (start_point, end_point), fill='white', width = 5)
dr.line(xy = ((img.width//2, 0),(img.width//2, img.height)), fill='white', width = 5)
img.save('img/generated.png')
```
